```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
bitcoin_df = pd.read_csv('../data/market_price.csv', names=['ds', 'y'])
bitcoin_df.head()
```

```python
bitcoin_df.drop(bitcoin_df.index[0], inplace=True)
bitcoin_df.head()
```

```python
bitcoin_df['y']=bitcoin_df['y'].astype(np.float)
bitcoin_df.info()
```

```python
bitcoin_test_df = bitcoin_df[-5:]
bitcoin_df = bitcoin_df.drop(bitcoin_df.index[-5:])
print(bitcoin_test_df)
print(bitcoin_df.tail())
```

```python
bitcoin_df['ds'] = pd.to_datetime(bitcoin_df['ds'])
bitcoin_df.index = bitcoin_df['ds']
bitcoin_df.set_index('ds', inplace=True)
bitcoin_df.head()
```

```python
bitcoin_df.plot()
plt.show()
```

```python
from pmdarima.arima import auto_arima

model_arima = auto_arima(bitcoin_df, trace=True, error_action='ignore',
                         start_p=1, start_q=1, max_p=3, max_q=3, 
                         suppress_warnings=True, stepwise=False, seasonal=False)

model_arima.fit(bitcoin_df)
```

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm

# AR = 2(p), 차분(d) = 1(d), MA = 2(q)
model = ARIMA(bitcoin_df, order=(1, 0, 1))

# trand : 상수 c(constant) / nc(no constant)
# disp  : 수렴정보를 나타냄
model_fit = model.fit(trand='c', disp=True, full_output=True)

# 학습 내용 summary
model_fit.summary()
```

```python
# plot_predict() : 학습 데이터만 가지고 예측한 결과
fig = model_fit.plot_predict()

# 잔차의 변동을 시각화
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
```

```python
# 학습 데이터셋으로부터 5일 뒤 예측
forecast_data = model_fit.forecast(steps=5)

# 마지막 5일의 예측 데이터
pred_y = forecast_data[0].tolist()

# 실제 5일의 데이터
test_y = bitcoin_test_df.y.values

# 마지막 5일의 예측 데이터 최소값
pred_y_lower = []

# 마지막 5일의 예측 데이터 최대값
pred_y_upper = []

for lower_upper in forecast_data[2]:
    lower = lower_upper[0]
    upper = lower_upper[1]
    pred_y_lower.append(lower)
    pred_y_upper.append(upper)

# 모델이 예상한 가격을 나타내는 그래프
plt.plot(pred_y, color='yellow')
# 모델이 예상한 가격의 최소값을 나타내는 그래프
plt.plot(pred_y_lower, color='red')
# 모델이 예상한 가격의 최대값을 나타내는 그래프
plt.plot(pred_y_upper, color='blue')
# 실제 가격을 나타내는 그래프
plt.plot(test_y, color='green')
```

```python
# 모델이 예상한 가격을 나타내는 그래프
plt.plot(pred_y, color='yellow')
# 실제 가격을 나타내는 그래프
plt.plot(test_y, color='green')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

# RMSE : 모델의 예측값과 실제 값의 차이를 하나의 숫자로 표현
#rmse = mean_squared_error(test_y, pred_y) ** 0.5
rmse = sqrt(mean_squared_error(pred_y, test_y))
rmse  # 점수가 낮아야 좋은 값
```

```python
from fbprophet import Prophet

bitcoin_df['ds'] = bitcoin_df.index

prophet = Prophet(seasonality_mode='multiplicative',
                  yearly_seasonality=True,
                  weekly_seasonality=True,
                  daily_seasonality=True,
                  changepoint_prior_scale=0.5)

# 모델학습
prophet.fit(bitcoin_df)
```

```python
# 5일을 내다보며 예측하는 작업
future_data = prophet.make_future_dataframe(periods=5, freq='d')
forecast_data = prophet.predict(future_data)
forecast_data.head()
```

```python
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# 모델링을 통해 만들어진 forcast_data 시각화
fig1 = prophet.plot(forecast_data)
```

```python
# 일, 주, 연 별
fig2 = prophet.plot_components(forecast_data)
```

```python
pred_y = forecast_data.yhat.values[-5:]
test_y = bitcoin_test_df.y.values
pred_y_lower = forecast_data.yhat_lower.values[-5:]
pred_y_upper = forecast_data.yhat_upper.values[-5:]


plt.plot(pred_y, color='gold')
plt.plot(pred_y_lower, color='red')
plt.plot(pred_y_upper, color='blue')
plt.plot(test_y, color='green')
```

```python
plt.plot(pred_y, color='gold')
plt.plot(test_y, color='green')
```

```python
y = bitcoin_df.y.values[5:]
y_pred = forecast_data.yhat.values[5:-5]
```

```python
rmse = sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)
print(rmse)
print(r2)
```
