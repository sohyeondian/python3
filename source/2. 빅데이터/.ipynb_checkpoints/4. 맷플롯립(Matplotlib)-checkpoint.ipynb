```python
# 차트 제목, 축(x, y) 제목
plt.title('BasicFlowChart')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
plt.title('BasicFlowChart')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'k')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
# 축 범위 axis() : [xmin, xmax, ymin, ymax]
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
plt.title('BasicFlowChart')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], color = '#e34152')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
# 축 범위 axis() : [xmin, xmax, ymin, ymax]
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
# 특정 영역에 색상으로 채워서 강조
# fill()
# fill_between() : x축 y축
# fill_betweenx()

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)

plt.fill_between(x[1:3], y[1:3], color = 'pink', alpha = 0.5)
# 네 점을 잇는 영역에 색을 채운다.
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)

plt.fill_between(x[1:3], y1[1:3], y2[1:3], 
                 color = 'lightgray', alpha = 0.5)
plt.show()
```

```python
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bo', t, t**3, 'g-.')
plt.show()
```

```python
# 격자 만들기
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'bo')
plt.plot(t, t**2, color = '#e35f62', marker = '*',
         linewidth = 2)
plt.plot(t, t**3, color = 'springgreen', marker = '^', markersize = 9)
plt.grid(True)

plt.show()
```

```python
# 가로 방향의 그리드 표시
# Grid(격자, 눈금자)
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'bo')
plt.plot(t, t**2, color = '#e35f62', marker = '*',
         linewidth = 2)
plt.plot(t, t**3, color = 'springgreen', marker = '^', markersize = 9)
# axis : both(default), x, y
plt.grid(True, axis = 'y', color = 'lightgray',
         alpha = 0.5, linestyle = '--')

plt.show()
```

```python
# Tick 축의 간격을 표시하는 눈금
# xticks(), yticks()
plt.title('Tick 사용하기')
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'bo')
plt.plot(t, t**2, color = '#e35f62', marker = '*',
         linewidth = 2)
plt.plot(t, t**3, color = 'springgreen', marker = '^', markersize = 9)

plt.xticks(np.arange(0, 2, 0.2), 
           labels = ['Jan', '', 'Feb', '','Mar', '', 'May', '', 'Jun',
                    '', 'July'])
plt.yticks(np.arange(0, 7), 
           ('0GB', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.show()
```
