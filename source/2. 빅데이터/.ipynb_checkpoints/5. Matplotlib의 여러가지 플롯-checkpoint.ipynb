```python
# 바 차트 : bar, barh(수평막대 horizontal)

y = [2, 3, 1]
x = np.arange(len(y))

xlabel = ['A', 'B', 'C']

plt.title('Bar 차트')

plt.xticks(x, xlabel)
plt.yticks(y)

plt.bar(x, y)

plt.xlabel('Word')
plt.ylabel('Fre')

plt.show()
```

```python
people = ['aaa', 'bbb', 'ccc', 'ddd']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title('BarH Chart')
plt.barh(y_pos, performance, xerr = error, alpha = 0.4)
plt.yticks(y_pos, people)
plt.xlabel('x-label')
plt.show()
```

```python
# pie : 파이 차트
labels = ['aaa', 'bbb', 'ccc', 'ddd']
size = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.title('Pie Chart')
plt.pie(size, labels = labels, colors = colors, explode = explode,
       startangle = 90, shadow = True, autopct = '%1.1f%%')

# 파이차트를 그릴 때 원의 형태가 유지될 수 있도록 설정
plt.axis('equal')

plt.show()
```

```python
# 히스토그램 : 빈도수, hist
# 데이터를 집계할 구간 - bins

x = np.random.randn(1000)

plt.title('Hist')
plt.hist(x, bins = 10)
plt.show()
```
