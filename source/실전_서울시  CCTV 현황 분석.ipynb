```python
# 서울시 자치 인구
pop = pd.read_excel('D:/python3/data/01. population_in_Seoul.xls', encoding='utf-8')
pop.head()
```

```python
pop = pd.read_excel('D:/python3/study/01. population_in_Seoul.xls',
                    header=2, # 2번째 행까지
                    usecols='B, D, G, J, N',  # 필요한 열 지정
                    encoding='utf-8')
pop.head()
```

```python
# 컬럼명 변경
pop.rename(columns={pop.columns[0]:'구별', 
                    pop.columns[1]:'인구수', 
                    pop.columns[2]:'한국인', 
                    pop.columns[3]:'외국인', 
                    pop.columns[4]:'고령자'}, inplace=True)
```

```python
pop.head()
```

```python
# 파생변수 : 최근 증가율
# 2013년도 이전의 데이터와 최근 3년간(14, 15, 16년) 증가율 계산
cctv['최근증가율'] = (cctv['2014년'] + cctv['2015년'] + cctv['2016년']) / cctv['2013년도 이전'] * 100
cctv.head()
```

```python
# 최근증가율을 기준으로 내림차순
cctv.sort_values(by='최근증가율', ascending=False).head()
```

```python
# cctv가 설치된 대 수(소계)를 기준으로 내림차순 정렬
cctv.sort_values(by='소계', ascending=False).head()
```

```python
# pop 데이터 셋의 첫행을 제거하는 작업
pop.drop([0], inplace=True)
pop.head()
```

```python
# 파생변수 : 외국인 비율과 고령자 비율
pop['외국인비율'] = pop['외국인'] / pop['인구수'] * 100
pop['고령자비율'] = pop['고령자'] / pop['인구수'] * 100
pop
```

```python
# 인구수를 기준으로 내림차순 정렬
pop.sort_values(by='인구수', ascending=False).head()
```

```python
# 외국인 수를 기준으로 내림차순
pop.sort_values(by='외국인', ascending=False).head()
```

```python
# 외국인 비율을 기준으로 내림차순
pop.sort_values(by='외국인비율', ascending=False).head()
```

```python
# 고령자비율을 기준으로 내림차순
pop.sort_values(by='고령자비율', ascending=False).head()
```

```python
# 두 개의 데이터프레임을 합치는 작업
df = pd.merge(cctv, pop, on='구별')
df.head()
```

```python
# cctv가 설치된 대수의 합 '소계'와 상관관계가 있는 피처를 찾아내는 작업
# 대체적으로 0.5 ~ 0.7 강한 상관관계
np.corrcoef(df['고령자비율'], df['소계'])
```

```python
# 인구수를 기준으로 내림차순 정렬
df.sort_values(by='인구수', ascending=False).head()
```

```python
df.set_index('구별', inplace=True)
df.head()
```

```python
df['소계'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.grid()
plt.show()
```

```python
# cctv와 인구수가 양의 상관관계를 가지고 있다는 것을 확인하여 산점도 그래프에 상관관계를 이용하는 선을 추가하는 작업
fp1 = np.polyfit(df['인구수'], df['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.plot(fx, f1(fx), ls ='dashed', lw = 3, color='g')
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.grid()
plt.show()
```

```python
# 오차를 계산한 후 오차를 기준으로 정렬, 구별 이름 추출
fp1 = np.polyfit(df['인구수'], df['소계'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

df['오차'] = np.abs(df['소계'] - df['인구수'])

df_sort = df.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14, 10))
plt.scatter(df['인구수'], df['소계'], c=df['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw = 3, color='g')

for i in range(10):
    plt.text(df_sort['인구수'][i]*1.02, df_sort['소계'][i]*0.98, df_sort.index[i], fontsize = 15)
    
plt.xlabel('인구수')
plt.ylabel('인구수당비율')
plt.grid()
plt.show()
```
