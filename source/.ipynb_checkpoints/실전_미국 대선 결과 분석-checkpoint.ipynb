### 데이터 분석하기 : 분석결과를 시각화하기
1. 각 후보 별 전체 지역 득표수를 계산하시오.


- candidate : 후보자
- votes : 득표수

```python
# ~ 별, ~ 계산
# 통계량을 계산할 컬럼명.groupby('그룹화할 기준 컬럼명')
# step1. condidate의 unique 확인
primary['candidate'].unique()
```

```python
# step2. 후보별 득표수의 합계
# step3 득표수 별 오름or내림차순 정렬
candidate_to_votes_s = primary.groupby('candidate')['votes'].sum().sort_values()
candidate_to_votes_s
```

```python
# step4. 후보별 전체지역득표수를 정렬한 결과를 이용하여 수평 막대 그래프
candidate_to_votes_s.plot(kind='barh', figsize=(10, 10))
plt.show()
```

2. 각 주별, 공화당과 민주당의 득표 비율을 계산하시오.


- state : 주
- party : 정당

```python
# 각 주별 => state 기준열1
# 공화당과 민주당 => party 기준열2
# 득표수 : votes
# 득표비율 = 해당득표수 / 전체득표수
state_party_votes_s = primary.groupby(['state', 'party'])['votes'].sum()
state_party_votes_s
```

```python
# 각 주별 득표수
state_votes_s = primary.groupby('state')['votes'].sum()
state_votes_s
```

```python
state_party_votes_p = state_party_votes_s / state_votes_s
state_party_votes_p
```

```python
state_party_votes_p.unstack().plot(kind='barh', stacked=True, 
                                   title='주별 정당 득표 비율', figsize=(10, 15))
plt.show()
```

3. 사용자 정의 함수를 활용하여 각 county별 당선된 후보의 백인유권자 비율을 계산하시오.


- fips : 유권자
- party : 정당
- condidate : 후보자
- RHI825214 : 백인비율

```python
# 득표수 1위 = 당선
func = lambda x : x.sort_values('votes', ascending=False).iloc[0]
```

```python
winners = primary.groupby('fips').agg(func)
winners
```

```python
counties['RHI825214'].head()
```

```python
winners_county_races = pd.merge(winners, counties[['fips', 'RHI825214']],
                               left_index=True, right_on='fips', how='left')
winners_county_races.head()
```

```python
winners_county_races = winners_county_races.rename(columns={'RHI825214': 'white_pcts'})
winners_county_races.head()
```

```python
# 정당&후보자 별 백인유권자 비율 평균
winners_county_white_pcts = winners_county_races.groupby(['party', 
                                                          'candidate'])['white_pcts'].mean()
winners_county_white_pcts
```

```python
winners_county_white_pcts.plot(kind='barh', fontsize=11, figsize=(7, 4)).set_xlim([50, 100])
plt.tight_layout()
plt.show()
```
