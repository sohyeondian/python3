### 당뇨병 환자 데이터셋(Diabetes)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load_ 데이터셋 이름()
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
print(diabetes.data.shape, diabetes.target.shape)
```

```python
diabetes.target[:3]
```

```python
plt.scatter(diabetes.data[:,2],diabetes.target)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 훈련데이터 생성
x = diabetes.data[:,2]
y = diabetes.target
```

```python
# step1 . w,b 라는 임의의 값을 정함
w = 1.0
b = 1.0
```

```python
#step2. 훈련데이터의 첫번째 샘플을 이용해서 예측값(y_hat)을 얻는다.
y_hat = x[0] * w + b
y_hat
```

```python
# step3. 타깃과 예측값을 비교한다.
y[0]
```

```python
# w값을 조절해서 예측값을 바꾸는 작업
# w 값을 0.1만큼 증가
w_inc = w + 0.1
y_hat_inc = x[0] * w_inc + b
y_hat_inc
```

```python
# w값을 조정한 후 예측값이 증가한 정도를 확인(변화율)
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
w_rate
```

```python
# 변화율을 업데이트 할 때 (가중치 w를 업데이트 하는 방법)
w_new  = w + w_rate
```

#### 변화율로 절편 업데이트 하기



```python
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
y_hat_inc
```

```python
b_rate = (y_hat_inc - y_hat) / (b_inc -b)
b_rate
```

```python
b_new = b + 1
b_new
```

### 오차 역전파로 가중치와 절편 업데이트하는 작업

```python
# 오차( y - y_hat)와 변화율을 곱해서 가중치(w)를 업데이터 하기
err = y[0] - y_hat
w_new = w + w_rate * err
b_new = b + 1 * err
print(w_new, b_new)
```

```python
y_hat = x[1] * w_new + b_new
err = y[1] - y_hat
w_rate = x[1]
w_new = w_new + w_rate * err
b_new = b_new + 1 * err

print(w_new, b_new)
```

```python
# zip() : 여러개의 배열에서 동시에 요소르 하나씩 반환해주는 함수
for x_i, y_i in zip(x,y):
    y_hat = x_i * w +b
    err = y_i - y_hat
    w_rate = x_i
    w = w + w_rate * err
    b = b + 1 * err
print(w, b)
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 에포크(epoch) : 입력 값에 대해 몇번이나 반복하여 실험했는지를 나타냄
# zip() : 여러개의 배열에서 동시에 요소르 하나씩 반환해주는 함수
for i in range(1,100):    
    for x_i, y_i in zip(x,y):
        y_hat = x_i * w + b
        err = y_i - y_hat
        w_rate = x_i
        w = w + w_rate * err
        b = b + 1 * err
print(w,b)
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 선형회귀의 뉴런 클래스

```python
class Neuron:
    def __init__(self):
        self.w = 1.0 # 가중치 초기화
        self.b = 1.0 # 절편 초기화
    def forpass(self, x): # 1차방정식을 계산하는 메서드
        y_hat = x * self.w + self.b
        return y_hat
    def backprob(self,x,err):
        w_gred = x * err # 가중치에 대한 경사하강법
        b_gred = 1 * err # 절편에 대한 경사하강법
        return w_gred, b_gred
    def fit(self, x, y, epochs=100):
        for i in range(epochs):
            for x_i, y_i in zip(x,y):
                y_hat = self.forpass(x_i) # 정방향 계산
                err = -(y_i - y_hat) # 오차 계산
                w_gred, b_gred =self.backprob(x_i, err) # 역방향
                self.w -= w_gred # 가중치 업데이트
                self.b -= b_gred # 절편 업데이트
```

```python
neuron = Neuron()
neuron.fit(x, y)
```

```python
plt.scatter(x, y)
plt1 = (-0.1, -0.1 * neuron.w + neuron.b)
plt2 = (0.15, 0.15 * neuron.w + neuron.b)

plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
