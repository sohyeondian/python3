```python
from tensorflow import keras

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

print(train_input.shape, train_target.shape)
#     (60000,28,28)(60000,)
print(test_input.shape , test_target.shape)
#     (10000,28,28)(10000,)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, axs = plt.subplots(1,10, figsize =(10,10))
for i in range(10):
    axs[i].imshow(train_input[i], cmap = 'gray_r')
    axs[i].axis('off')
plt.show()
```

```python
print([train_target[i] for i in range(10)])
```

```python
import numpy as np
print(np.unique(train_target, return_counts= True))
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_validate
train_sclaed = train_input / 255.0   # 0~1 사이의 정규화 값을 바꾼다.
train_scaled = train_sclaed.reshape(-1,28*28)
print(train_scaled.shape)

sc = SGDClassifier(loss= 'log', max_iter = 5, random_state =42) # max_iter = epoch (반복횟수
scores = cross_validate(sc, train_scaled, train_target, n_jobs = -1) # cross_validate (교차검증)
print(np.mean(scores['test_score']))
```

```python
#10개의 샘플마다 방정식
#시그모이드 함수 사용 , 손실로
# 인공신경망이 아닌  unit 이라는 표현으로 많이쓴다.
# 퍼셉트론의 형태를 가지고 있다.
# 1.0 텐서플로 줄여놓은게 케라스
# 2.0 이후로는 텐서플로 안에 케라스가 들어감
# 하나의 데이터로만은 정확한 결과가 안나오기 때문에 검증데이타 셋으로  분별한다.
# Dense 는 케라스 레이어라는 배이어 층의 역활 층이가지고 있는 역활, 속성
```

```python
from sklearn.model_selection import train_test_split
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled
                                                                      , train_target, test_size = 0.2, random_state =41)
print(train_scaled.shape, train_target.shape)
print(val_scaled.shape, val_target.shape)
dense = keras.layers.Dense(10,activation= 'softmax', input_shape=(784,))
model = keras.Sequential(dense)
```

```python
model.compile(loss = 'sparse_categorical_crossentropy',metrics= ['accuracy'])
```

```python
model.fit(train_scaled, train_target, epochs = 5)
```

```python
model.evaluate(val_scaled, val_target)
```
