### k-mean 클러스터링
- 분류(Classifier)와 군집(Clustering)은 다르다
- 비지도 학습(Unsupervised Learning) 알고리즘
- k개로 군집화(cluster) == class
- mean : 각 클러스터의 중심과 데이터들의 평균 거리

#### k-mean 클러스터링 과정
  1. 데이터
  2. 몇 개의 클러스터를 만들 지 결정
  3. 클러스터링할때, 클러스터의 중심(centeroid) 설정

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
%matplotlib inline

from sklearn.cluster import KMeans
```

```python
# 데이터 생성
df = pd.DataFrame(columns=['x', 'y'])
```

```python
for i in range(30):
    df.loc[i] = [random.randrange(1, 100),
                 random.randrange(1, 100)]
```

```python
df
```

```python
import seaborn as sns
sns.lmplot(x='x', y='y', data=df, fit_reg=False, scatter_kws={'s':30})
plt.title('K-means')
plt.show()
```

```python
# 데이터를 numpy 배열로
points = df.values
```

```python
# k-means 알고리즘 3개의 cluster 생성
kmeans = KMeans(n_clusters=3).fit(points)
```

```python
# 중심값 생성
kmeans.cluster_centers_
```

```python
df['cluster'] = kmeans.labels_
df
```
