```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(6, 4))
df
```

```python
df.columns = ['A', 'B', 'C', 'D']

# pandas가 제공하는 날짜 형식 데이터
df.index = pd.date_range('20210501', periods=6)

df
```

### 결측치(NaN)

```python
df['E'] = [1.0, np.nan, 3.5, 6.1, np.nan, 7.0]
df
```

```python
df.dropna(how='all')  # 모든 값이 결측값인 행 제거 - 현재 제거할 대상 없음
```

```python
df.dropna(how='any')  # 결측값이 있는 행 제거
```

```python
# fillna(value=)
df.fillna(5.0)
```

```python
# isnull() : 결측값입니까? True or False
df.isnull()
```

```python
# 결측값을 가지고 있는 로우만 추출
df.loc[df.isnull()['E'], :]
```
