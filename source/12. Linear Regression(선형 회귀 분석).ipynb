```python
# 피처의 정규화 or 스케일링
scale_columns = ['승', '패', '세', '홀드', '블론', '경기', '선발', '이닝', '삼진/9',
                 '볼넷/9', '홈런/9', 'BABIP', 'LOB%', 'ERA', 'RA9-WAR', 'FIP', 
                 'kFIP', 'WAR', '연봉(2017)']
```

### 정규화 작업
- 최소,최대값 정규화(min_max)
- 평균과 표준편차를 이용한 정규화(z-value)

```python
def standard_scaling(df, columns):
    for col in columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x : (x - series_mean) / series_std)
    return df
```

```python
picher_df = standard_scaling(picher, scale_columns)
picher_df.head()
```

```python
picher_df = picher_df.rename(columns={'연봉(2018)':'y'})
```

```python
picher_df.head()
```

```python
# 범주형 데이터 '팀명'을 정규화 : 원핫인코딩 - get_dummies()
team_scaling = pd.get_dummies(picher_df['팀명'])
picher_df = picher_df.drop('팀명', axis=1)
picher_df = picher_df.join(team_scaling)
team_scaling.head()
```

### Train data, Test data 분리

```python
picher_df.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

# 데이터는 대문자 X, 레이블은 소문자 y
y = picher_df['y']
# 선수명, 'y'를 뺀 나머지는 데이터다
X = picher_df[picher_df.columns.difference(['선수명', 'y'])]

# train:test = 8:2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
X_train.shape
```

모델을 생성하고 학습

```python
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)
```

- 선형회귀분석이라는 것은 선형(직선)을 만들어내는 것이다.
- 회귀방정식을 만들어 새로운 값이 들어왔을 때 예측
- statesmodel 라이브러리에 OLS클래스

```python
import statsmodels.api as sm

X_train = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train).fit()
# model = sm.OLS.from_formula('y_train ~ X_train', data=picher_df)
model.summary()
```

```python
# 회귀 계수를 이용해서 시각화
plt.rcParams['figure.figsize'] = [20, 16]
coefs = model.params.tolist()
coefs_series = pd.Series(coefs)

x_labels = model.params.index.tolist()

ax = coefs_series.plot(kind='bar')
ax.set_ylabel('coef')
ax.set_xlabel('x_feature')
ax.set_xticklabels(x_labels)
plt.show()
```

##### 예측 모델 평가

```python
# 데이터는 대문자 X, 레이블은 소문자 y
y = picher_df['y']
# 선수명, 'y'를 뺀 나머지는 데이터다
X = picher_df[picher_df.columns.difference(['선수명', 'y'])]

# train:test = 8:2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)

# 회귀모델 생성하고 학습
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)

# 회귀분석모델 평가
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

### 회귀모델을 평가하는 RMSE score
* 이 값이 높을수록 모델의 예측력이 부정확하다는 것을 의미

```python
from math import sqrt
from sklearn.metrics import mean_squared_error

# train data의 RMSE score
y_prediction = lr.predict(X_train)
print(sqrt(mean_squared_error(y_train, y_prediction)))

# test data의 RMSE score
y_prediction = lr.predict(X_test)
print(sqrt(mean_squared_error(y_test, y_prediction)))
```

### 다중공선성
- 변수간 상관 관계가 높아 분석에 부정적인 영향을 미치는 것
- 분산팽창요인(VIF)이라는 계수로 평가
- VIF 계수가 10~15 정도를 넘으면 그 피처는 다중공선성의 문제가 발생했다고 판단
- 다중공선성의 문제를 해결하는 방법: 해당 피처를 drop
- 안전 : VIF < 5, 주의 : 5 < VIF < 10, 위험 : 10 < VIF

```python
# 회귀분석의 예측 성능을 높이기 위한 방법 : 다중공선성확인
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = picher_df[picher_df.columns.difference(['선수명', 'y'])]
y = picher_df['y']

vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['feature'] = X.columns
vif.round(1)
```

```python
# 적절한 피처를 선택해서 회귀 분석 수행
X = picher_df[['FIP', 'WAR', '볼넷/9', '삼진/9', '연봉(2017)']]
y = picher_df['y']

# train:test = 8:2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    random_state=19)
```

```python
# 회귀모델 생성하고 학습
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)
```

```python
# 회귀분석모델 평가
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
# VIF 계수 출력
X = picher_df[['FIP', 'WAR', '볼넷/9', '삼진/9', '연봉(2017)']]
vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['feature'] = X.columns
vif.round(1)
```

### 다중 공선성을 제거하기 위한 피처 선별

2. 단, 유사한 feature의 경우 둘 중 1개만 제거(FIP, kFIP의 경우)
3. 제거 후 VIF 계수 재출력

6. feature의 p-value 검정
