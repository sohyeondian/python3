### 타이타닉의 생존자 정보를 제공하는 데이터셋

````python
# 라이브러리 로딩
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns
````

````python
titanic = sns.load_dataset('titanic')
titanic.head()
````

### 타이타닉의 컬럼 설명
- survived: 생존여부 0-사망, 1-생존
- pclass: 고객 등급
- sex: 성별, male-남, femail-여
- sibsp: 동반한 형제 혹은 자매의 수
- parch: 동방한 부모 혹은 자녀의 수 
- embarked : 승선한 항구의 이니셜
- class: 객실 등급 1-2-3
- who: 대표 고객의 성별 man woman
- adult_male: 성인남성여부 True, False
- deck: 갑판
- embark_town: 승선한 항구의 지역명
- alive: 생존 여부 yes, no
- alone: 혼자 True, False

````python
import missingno as msno

msno.matrix(titanic)
plt.show()
````

````python
# 결측된 데이터를 처리하는 방법
# 결측된 데이터가 너무 많을 경우 해당 열 전체 삭제
# 결측된 데이터가 일부일 경우 결측값을 대체하는 작업
````

````python
# 데이터가 절반 이상이 없는 열을 삭제
titanic = titanic.dropna(thresh=int(len(titanic)*0.5), axis=1)
msno.matrix(titanic)
plt.show()
````

````python
sns.countplot(titanic.embark_town)
plt.show()
````

````python
sns.kdeplot(titanic.age)
plt.show()
````

### 결측데이터를 다른 값으로 대체하는 작업
- 해당 열의 비결측 데이터의 평균값(mean) 혹은 중앙값(median) 등으로 대체하는 작업
- scikit-learn의 SimpleImputer 클래스를 활용
- strategy 옵션에 'mean', 'median', 'most_frequent'
     - 데이터가 실수 연속 값인 경우 평균 or 중앙값 사용
     - 값의 분포가 대칭적이면 평균, 값이 분포가 심하게 비대칭적이면 중앙값
     - 데이터가 범주형이거나 정수값인 경우 최빈값
- fit_transform 메서드를 활용하여 대체값으로 채워진 데이터프레임 생성

````python
titanic.isnull().sum()
````

````python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')
````

````python
# embarked & embark_town : 문자열 -> 최빈값
titanic['embark_town'] = imputer.fit_transform(titanic[['embark_town']])
titanic['embarked'] = imputer.fit_transform(titanic[['embarked']])
````

````python
msno.matrix(titanic)
plt.show()
````

````python
# age : 비대칭적인 숫자 -> 중앙값
imputer_age = SimpleImputer(strategy='median')
titanic['age'] = imputer.fit_transform(titanic[['age']])
msno.matrix(titanic)
plt.show()
````

````python
df_titanic = titanic.drop(['class', 'alive', 'who', 'embarked', 'embark_town'], axis=1)
df_titanic.head(10)
````

````python
# 등급별 고객정보, 각 객체 별 건수
df_titanic['pclass'].value_counts()
````

````python
df_titanic['survived'].value_counts().plot.bar()
````

````python
ax = sns.countplot(x='pclass', hue='survived', data=df_titanic)
````

````python
# 타이타닉 생존력이 가장 높은 집단 찾아보기
df_titanic.groupby('sex')['survived'].value_counts().unstack().plot.bar() 
# 여성 생존률이 더 높음
````

````python
df_titanic.groupby('adult_male')['survived'].value_counts().unstack().plot.bar()
````

````python
df_titanic.groupby(['sex', 'pclass'])['survived'].value_counts().unstack().plot(kind='barh', 
                                                                                stacked=True)
plt.show()
````

````python
df = df_titanic.groupby(['age','pclass'])['survived'].value_counts()
sns.kdeplot(df)
plt.show()
````
