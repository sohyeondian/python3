```python
# 매직명령어
%matplotlib inline

# 필요한 라이브러리 로딩
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 로딩 : read_csv(파일명, sep, encoding='utf-8')
cctv = pd.read_csv('D:/Data Science/data/CCTV_in_Seoul.csv', encoding='utf-8')
```

```python
# 데이터 확인 : head(), tail()
cctv.head()
```

```python
# 컬럼명
cctv.columns
```

```python
# 피처(컬럼)명 변경
# 컬럼명 변경 : rename
# cctv.columns[0]
# inplace = True : 변경 내용을 원본에 바로 갱신하기 위한 옵션
cctv.rename(columns={cctv.columns[0]:'구별'}, inplace = True)
cctv.head()
```

```python
# 서울시 인구 현황 데이터 로딩
pop = pd.read_excel('D:/Data Science/data/01. population_in_Seoul.xls', encoding='utf-8')
pop.head()
```

```python
pop = pd.read_excel('D:/Data Science/data/01. population_in_Seoul.xls',
                    header = 2,
                    usecols = 'B, D, G, J, N',
                    encoding='utf-8')
pop.head()
```

```python
# 컬럼명 변경
pop.rename(columns={pop.columns[0]:'구별',
                   pop.columns[1]:'인구수',
                   pop.columns[2]:'한국인',
                   pop.columns[3]:'외국인',
                   pop.columns[4]:'고령자'}, inplace=True)
pop.head()
```

```python
# 파생변수 : 최근증가율
# 2013년도 이전에 데이터와 최근 3년간(14, 15, 16년) 증가율을 계산
cctv['최근증가율'] = (cctv['2014년'] + cctv['2015년'] + cctv['2016년']) / cctv['2013년도 이전'] * 100
cctv.head()
```

```python
# 최근증가율을 기준으로 내림차순
cctv.sort_values(by='최근증가율', ascending=False).head()
```

```python
# cctv가 설치된 대수(소계)를 기준으로 내림차순 정렬
cctv.sort_values(by='소계', ascending=False).head()
```

```python
# pop 데이터셋의 첫행을 제거하는 작업
pop.drop([0], inplace=True)
pop.head()
```

```python
# unique() : 중복값을 제거하고 고유값을 확인하는 함수
pop['구별'].unique()
```

```python
# 결측치 확인 : [], isnull()
pop[pop['구별'].isnull()]
```

```python
# 결측치 제거
pop.drop([26], inplace=True)
pop
```

```python
# 파생변수 : 외국인비율과 고령자비율
pop['외국인비율'] = pop['외국인'] / pop['인구수'] * 100
pop['고령자비율'] = pop['고령자'] / pop['인구수'] * 100
pop
```

```python
# 인구수를 기준으로 내림차순 정렬
pop.sort_values(by='인구수', ascending=False).head()
```

```python
# 외국인 수를 기준으로 내림차순
pop.sort_values(by='외국인', ascending=False).head()
```

```python
# 외국인 비율을 기준으로 내림차순 
pop.sort_values(by='외국인비율', ascending=False).head()
```

```python
# 고령자 수를 기준으로 내림차순
pop.sort_values(by='고령자', ascending=False).head()
```

```python
# 고령자비율 기준으로 내림차순
pop.sort_values(by='고령자비율', ascending=False).head()
```

```python
# 두 개의 데이터프레임을 합치는 작업
df = pd.merge(cctv, pop, on = '구별')
df.head()
```

```python
# cctv가 설치된 대수의 합 '소계'와 상관관계가 있는 피처를 찾아내는 작업
np.corrcoef(df['고령자비율'], df['소계'])
```

```python
np.corrcoef(df['외국인비율'], df['소계'])
```

```python
np.corrcoef(df['인구수'], df['소계'])
```

```python
# 인구수를 기준으로 내림차순 정렬
df.sort_values(by='인구수', ascending=False).head()
```

```python
# 분석에 불피요하다 판단되는 피처 제거하는 작업
del df['2013년도 이전']
del df['2014년']
del df['2015년']
del df['2016년']
df.head()
```

```python
# '구별' 컬럼을 index로 설정하는 작업
df.set_index('구별', inplace=True)
df.head()
```

```python
df['소계'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
# 인구대비 CCTV 비율
df['cctv비율'] = df['소계'] / df['인구수'] * 100
df['cctv비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
# cctv와 인구수가 양의 상관관계를 가지고 있다는 것을 확인하여 산점도 그래프에
# 상관관계를 이용하는 선을 추가하는 작업
fp1 = np.polyfit(df['인구수'], df['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
# 오차를 계산한 후 오차를 기준으로 정렬, 구별 이름 추출
fp1 = np.polyfit(df['인구수'], df['소계'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

df['오차'] = np.abs(df['소계'] - f1(df['인구수']))

df_sort = df.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14, 10))
plt.scatter(df['인구수'], df['소계'], c=df['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, 
             df_sort['소계'][n]*0.98, df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구수당비율')
plt.grid()
plt.show()
```
